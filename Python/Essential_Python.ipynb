```
# my_phone_tuple[2]= 4 
# This operation is not supported as tuples are immutable

# Question :
# WHY? What could be the advantage of being immutable?  

# Experiment : Comparing execution time in lists and tuples

print("----Comparing execution time for an element search in lists & tuples----")

import time
my_list = list(range(100000))
tic = time.time()
my_list_squared = list(map(lambda x:x**2, my_list))
toc = time.time()
print("Elapsed time in seconds for creating a list",toc-tic,"seconds")

my_tuple = tuple(range(100000))
tic = time.time()
my_tuple_squared = tuple(map(lambda x:x**2, my_tuple))
toc = time.time()
print("Elapsed time in seconds for creating a tuple",toc-tic,"seconds" )

# Answer: 
# Faster iteration over large amounts of data in tuples(because of immutability)
```

# Section 4 : Defining Functions

```
def sum_of_squares(stop_val,start_val=0, ):
  sum_count = 0
  for i in range(start_val ,stop_val):
    sum_count += i**2
  return sum_count

s = sum_of_squares(10,1)
print(s)  
```

# Section 5 : Python Packages

```
# Importing standard python packages
import math
import time

a = math.sqrt(100)
print(a)

# An optimized way of importing packages
import math as m            # abrrevation for package name mitigates rewriting 
from math import factorial  # import a particular function from the package
```

```
# Why do we need packages like math ? 
# To avoid rewriting code and to use an existing optimized code(computationally)

# Custom function for factorial
def cust_factorial(num):
  product = 1
  for i in range(1,num+1):
    product *= i 
  return product
```

```
# Experiment: Find factorial of list of numbers (custom func. Vs. package func.)
num_list = list(range(1,100))

tic = time.time()
entry_cust_factorial = list()
for entry in num_list:
  entry_cust_factorial.append(cust_factorial(entry))
toc = time.time()
print("Elapsed time for executing the custom function is ",toc-tic,"seconds")

tic = time.time()
entry_factorial = list()
for entry in num_list:
  entry_factorial.append(factorial(entry))
toc = time.time()
print("Elapsed time for executing the package function is ",toc-tic,"seconds")
```

# Section 7 : Class
**OOP with Python**

```
class MobilePhone : 
  """This class contains basic attributes for a mobile phone"""
  def __init__(self,name,isAndroid = True,screenSize = 5,rating = -1 ):
    self.name = name
    self.isAnrdoid = isAndroid
    self.screenSize = screenSize
    self.rating = rating
  
  def has_rating(self):
    return self.rating > -1

new_phone = MobilePhone('iPhone 5S',False,4.3,5) 
type(new_phone)
```

```
print(new_phone.name,new_phone.isAnrdoid,new_phone.screenSize,new_phone.rating)
print(new_phone.has_rating())

# Docstring can be accessed as follows
MobilePhone.__doc__
```

```
# Inherting an existing class
class iphone(MobilePhone):
  def __init__(self,name):
    MobilePhone.__init__(self,name,False,4)

phone = iphone('iphone 5S')
print(phone.isAnrdoid,phone.screenSize)
  
```

```
class iphone(MobilePhone):
  def __init__(self,name): # ___init__ is an in-built function to initialize
    MobilePhone.__init__(self,name,False,4)

  def __str__(self): # __str__ is an in-built function to print strings
    return self.name + " " + str(self.isAnrdoid) + " " + str(self.screenSize)

phone = iphone('iphone 5S')
print(phone) # prints all attributes of a class
```

# Section 8 : The Numpy Library

```
import numpy as np
X = np.array([1,2,3])
print(X)
type(X)
```

```
# Convert a given list to np array
L = [1,2,3,4]
X = np.asarray(L)
type(X)
print(X)
```

```
# Convert elements in a numpy array to other data-type
X = np.asarray(X,float)
print(X)
```

```
# Printing the shape of an array
print(X.shape) # shape is an attribute of a numpy array
```

```
# 2D arrays 
A = np.array([[1,2,3,4],[5,6,7,8]])
print(A)
print(A.shape) # rows and columns
```

```
print(A[1,1])   # prints a particular entry
print(A[0,0:2]) # prints elements in a range
print(A[0,:])   # prints entire row
```

```
# In-built arrays

Z = np.zeros((5,5)) # Zero matrix 
I = np.eye(3,3) # Identity matrix     
R = np.random.random((4,4)) # Random number
print(Z,"\n","\n",I,"\n","\n",R) 
```

```
# Operations on np arrays
Transp = R.T
print(Transp)
Reshap = Transp.reshape(16,1)
print("\n",Reshap) #Important for Deep Learning
```

```
# Arithematic operations in np arrays
A = np.array([[1,2,3],[4,5,6]])
B = np.array([[1,2,3],[4,5,6]])
C = A + B
D = A - B
E = A*B
print(C,"\n","\n",D,"\n","\n",E)

# Brodcasting in np arrays
print("\n","\n",A+1,"\n","\n",A-1,"\n","\n",A*2,"\n","\n",A**2,"\n","\n",A/2)
```

```
M = np.floor(np.random.random((2,3))*10)
print(M)
```

```
u = [1,2,3,]
v = [1,0,-1]

p1 = np.inner(u,v)
p2 = np.outer(u,v)
print(p1,"\n",p2)
```

```
A = np.ones((2,3))
B = np.ones((3,2))

print(np.dot(A,B))
print("\n", np.dot(B.T,A.T))
```

```
# In-built Numpy functions
# Sum 
print(A.sum())       # sum of all elements
print(A.sum(axis=0)) # column-wise sum
print(A.sum(axis=1)) # row-wise sum

# Max 
print(A.max())
print(A.max(axis=0))
print(A.max(axis=1))

# More functions on : - https://numpy.org/devdocs/user/absolute_beginners.html
```
